**Grayscale Image processing**

```
import numpy as np
import cv2
import math
from google.colab.patches import cv2_imshow
path = '/content/lena.jpg'
#https://classroom.google.com/c/NzA3ODQ5MzU1MjUw?cjc=qmclhg7
img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
cv2_imshow(img)

print(img.shape[0])
print(img.shape[1])
print(img.max())
print(img.min())

for i in range(img.shape[0]):
  for j in range(img.shape[1]):
    a = img.item(i,j)
    if a > 155:
      img.itemset((i,j),255)
    else:
      img.itemset((i,j),0)

cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

**Grayscale Image Histogram Visualization**

```
from google.colab.patches import cv2_imshow


import cv2
import math
import matplotlib.pyplot as plt

img = cv2.imread("/content/lena.jpg",cv2.IMREAD_GRAYSCALE)
cv2_imshow(img)
cv2.waitKey(0)

print(img.shape,img.max(),img.min())

def histShow(img):

    arr = []
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
          arr.append(img[i][j])


    plt.hist(arr)
    plt.xlabel('values')
    plt.ylabel('count')

    plt.show()

histShow(img)
```

**Apple and Orange Mix**

```
from google.colab.patches import cv2_imshow


import cv2
import math
import matplotlib.pyplot as plt
import numpy



app=cv2.imread("/content/apple.jpg",cv2.IMREAD_GRAYSCALE)
org=cv2.imread("/content/orange.jpg",cv2.IMREAD_GRAYSCALE)

mix=app.copy()

w=app.shape[1]/2

for x in range(app.shape[0]):
    for y in range(app.shape[1]):
      mix.itemset((x,y),(0.5*app[x][y])+(0.5*org[x][y]))

cv2_imshow(app)
cv2_imshow(org)
cv2_imshow(mix)
#mix.save('merged_image.jpg')

cv2.waitKey(0)
```

**Lenas Negative Image Create**

```
from google.colab.patches import cv2_imshow
import cv2
import math
import matplotlib.pyplot as plt
import numpy

img=cv2.imread("/content/lena.jpg",cv2.IMREAD_GRAYSCALE)
cv2_imshow(img)
cv2.waitKey(0)
max=256
for x in range(img.shape[0]):
    for y in range(img.shape[1]):
      img.itemset((x,y),max-1-img[x][y])

cv2_imshow(img)
cv2.waitKey(0)
```

**High Contrast Transformed**

```
from google.colab.patches import cv2_imshow
import cv2
import math
import matplotlib.pyplot as plt
import numpy

img=cv2.imread("/content/lena.jpg",cv2.IMREAD_GRAYSCALE)
cv2_imshow(img)
cv2.waitKey(0)
c=10
for x in range(img.shape[0]):
    for y in range(img.shape[1]):
      img.itemset((x,y),c*math.log(img[x][y]+1))

cv2_imshow(img)
cv2.waitKey(0)
```

**Low Constrat (Gamma Transformation)**

```
from google.colab.patches import cv2_imshow
import cv2
import math
import matplotlib.pyplot as plt
import numpy

img=cv2.imread("/content/lena.jpg",cv2.IMREAD_GRAYSCALE)
cv2_imshow(img)
cv2.waitKey(0)
gama=0.3
c=256/(256**gama)
for x in range(img.shape[0]):
    for y in range(img.shape[1]):
      img.itemset((x,y),c*(img[x][y]**gama))

cv2_imshow(img)
cv2.waitKey(0)
```
